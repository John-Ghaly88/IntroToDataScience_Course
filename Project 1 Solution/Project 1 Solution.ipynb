```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

##Extracting the data:
#Reading csv files
df_train = pd.read_csv(r'Train.csv')
df_test = pd.read_csv(r'Test.csv')


#Concatinating the 2 tables
df_temp = [df_train, df_test]
df = pd.concat(df_temp)


#Columns' variables to work on
pl = df.PetalLengthCm
pw = df.PetalWidthCm



##Equal Interval discretization with n = 2 intervals:
#Discretizing values
i2_dpl = pd.cut(pl, bins=2, labels=[1,2])
i2_dpw = pd.cut(pw, bins=2, labels=[1,2])


#Calculating the average
i2_c_vc = 0
i2_c_vi = 0
i2_c_s = 0

i2_spl_vc = 0
i2_spl_vi = 0
i2_spl_s = 0

i2_spw_vc = 0
i2_spw_vi = 0
i2_spw_s = 0

for i in range(0, 105):
    if df.Species.values[i] == 'Iris-versicolor':
        i2_c_vc = i2_c_vc + 1
        i2_spl_vc = i2_spl_vc + i2_dpl.values[i]
        i2_spw_vc = i2_spw_vc + i2_dpl.values[i]
        
    elif df.Species.values[i] == 'Iris-virginica':
        i2_c_vi = i2_c_vi + 1
        i2_spl_vi = i2_spl_vi + i2_dpl.values[i]
        i2_spw_vi = i2_spw_vi + i2_dpl.values[i]
    
    elif df.Species.values[i] == 'Iris-setosa':
        i2_c_s = i2_c_s + 1
        i2_spl_s = i2_spl_s + i2_dpl.values[i]
        i2_spw_s = i2_spw_s + i2_dpl.values[i]
        
i2_apl_vc = i2_spl_vc / i2_c_vc
i2_apl_vi = i2_spl_vi / i2_c_vi
i2_apl_s = i2_spl_s / i2_c_s

i2_apw_vc = i2_spw_vc / i2_c_vc
i2_apw_vi = i2_spw_vi / i2_c_vi
i2_apw_s = i2_spw_s / i2_c_vc


#The Euclidean distance
i2_ed = ['null'] * 45
j1 = 0

for i in range(105, 150):
    #d = ((((x2 - x1 )**2) + ((y2-y1)**2) )**0.5)
    i2_d_vc = ((((i2_apl_vc - i2_dpl.values[i] )**2) + ((i2_apw_vc - i2_dpw.values[i])**2) )**0.5)
    i2_d_vi = ((((i2_apl_vi - i2_dpl.values[i] )**2) + ((i2_apw_vi - i2_dpw.values[i])**2) )**0.5)
    i2_d_s = ((((i2_apl_s - i2_dpl.values[i] )**2) + ((i2_apw_s - i2_dpw.values[i])**2) )**0.5)
    
    if (i2_d_vc < i2_d_vi) and (i2_d_vc < i2_d_s):
        i2_ed[j1] = 'Iris-versicolor'
        
    elif (i2_d_vi < i2_d_vc) and (i2_d_vi < i2_d_s):
        i2_ed[j1] = 'Iris-virginica'
        
    elif (i2_d_s < i2_d_vc) and (i2_d_s < i2_d_vi):
        i2_ed[j1] = 'Iris-setosa'
    j1 = j1 + 1


#Accuracy
i2_c = 0
for i in range(0, 45):
    if df_test.Species.values[i] == i2_ed[i]:
        i2_c = i2_c + 1
i2_a = (i2_c / 45) * 100



##Equal Interval discretization with n = 3 intervals
#Discretizing values
i3_dpl = pd.cut(pl, bins=3, labels=[1,2,3])
i3_dpw = pd.cut(pw, bins=3, labels=[1,2,3])


#Calculating the average
i3_c_vc = 0
i3_c_vi = 0
i3_c_s = 0

i3_spl_vc = 0
i3_spl_vi = 0
i3_spl_s = 0

i3_spw_vc = 0
i3_spw_vi = 0
i3_spw_s = 0

for i in range(0, 105):
    if df.Species.values[i] == 'Iris-versicolor':
        i3_c_vc = i3_c_vc + 1
        i3_spl_vc = i3_spl_vc + i3_dpl.values[i]
        i3_spw_vc = i3_spw_vc + i3_dpl.values[i]
        
    elif df.Species.values[i] == 'Iris-virginica':
        i3_c_vi = i3_c_vi + 1
        i3_spl_vi = i3_spl_vi + i3_dpl.values[i]
        i3_spw_vi = i3_spw_vi + i3_dpl.values[i]
    
    elif df.Species.values[i] == 'Iris-setosa':
        i3_c_s = i3_c_s + 1
        i3_spl_s = i3_spl_s + i3_dpl.values[i]
        i3_spw_s = i3_spw_s + i3_dpl.values[i]
        
i3_apl_vc = i3_spl_vc / i3_c_vc
i3_apl_vi = i3_spl_vi / i3_c_vi
i3_apl_s = i3_spl_s / i3_c_s

i3_apw_vc = i3_spw_vc / i3_c_vc
i3_apw_vi = i3_spw_vi / i3_c_vi
i3_apw_s = i3_spw_s / i3_c_vc


#The Euclidean distance
i3_ed = ['null'] * 45
j2 = 0

for i in range(105, 150):
    #d = ((((x2 - x1 )**2) + ((y2-y1)**2) )**0.5)
    i3_d_vc = ((((i3_apl_vc - i3_dpl.values[i] )**2) + ((i3_apw_vc - i3_dpw.values[i])**2) )**0.5)
    i3_d_vi = ((((i3_apl_vi - i3_dpl.values[i] )**2) + ((i3_apw_vi - i3_dpw.values[i])**2) )**0.5)
    i3_d_s = ((((i3_apl_s - i3_dpl.values[i] )**2) + ((i3_apw_s - i3_dpw.values[i])**2) )**0.5)
    
    if (i3_d_vc < i3_d_vi) and (i3_d_vc < i3_d_s):
        i3_ed[j2] = 'Iris-versicolor'
        
    elif (i3_d_vi < i3_d_vc) and (i3_d_vi < i3_d_s):
        i3_ed[j2] = 'Iris-virginica'
        
    elif (i3_d_s < i3_d_vc) and (i3_d_s < i3_d_vi):
        i3_ed[j2] = 'Iris-setosa'
    j2 = j2 + 1


#Accuracy
i3_c = 0
for i in range(0, 45):
    if df_test.Species.values[i] == i3_ed[i]:
        i3_c = i3_c + 1
i3_a = (i3_c / 45) * 100



##Equal Frequency discretization with n = 2 intervals
#Discretizing values
f2_dpl = pd.qcut(pl, q=2, labels=[1,2])
f2_dpw = pd.qcut(pw, q=2, labels=[1,2])


#Calculating the average
f2_c_vc = 0
f2_c_vi = 0
f2_c_s = 0

f2_spl_vc = 0
f2_spl_vi = 0
f2_spl_s = 0

f2_spw_vc = 0
f2_spw_vi = 0
f2_spw_s = 0

for i in range(0, 105):
    if df.Species.values[i] == 'Iris-versicolor':
        f2_c_vc = f2_c_vc + 1
        f2_spl_vc = f2_spl_vc + f2_dpl.values[i]
        f2_spw_vc = f2_spw_vc + f2_dpl.values[i]
        
    elif df.Species.values[i] == 'Iris-virginica':
        f2_c_vi = f2_c_vi + 1
        f2_spl_vi = f2_spl_vi + f2_dpl.values[i]
        f2_spw_vi = f2_spw_vi + f2_dpl.values[i]
    
    elif df.Species.values[i] == 'Iris-setosa':
        f2_c_s = f2_c_s + 1
        f2_spl_s = f2_spl_s + f2_dpl.values[i]
        f2_spw_s = f2_spw_s + f2_dpl.values[i]
        
f2_apl_vc = f2_spl_vc / f2_c_vc
f2_apl_vi = f2_spl_vi / f2_c_vi
f2_apl_s = f2_spl_s / f2_c_s

f2_apw_vc = f2_spw_vc / f2_c_vc
f2_apw_vi = f2_spw_vi / f2_c_vi
f2_apw_s = f2_spw_s / f2_c_vc


#The Euclidean distance
f2_ed = ['null'] * 45
j3 = 0

for i in range(105, 150):
    #d = ((((x2 - x1 )**2) + ((y2-y1)**2) )**0.5)
    f2_d_vc = ((((f2_apl_vc - f2_dpl.values[i] )**2) + ((f2_apw_vc - f2_dpw.values[i])**2) )**0.5)
    f2_d_vi = ((((f2_apl_vi - f2_dpl.values[i] )**2) + ((f2_apw_vi - f2_dpw.values[i])**2) )**0.5)
    f2_d_s = ((((f2_apl_s - f2_dpl.values[i] )**2) + ((f2_apw_s - f2_dpw.values[i])**2) )**0.5)
    
    if (f2_d_vc < f2_d_vi) and (f2_d_vc < f2_d_s):
        f2_ed[j3] = 'Iris-versicolor'
        
    elif (f2_d_vi < f2_d_vc) and (f2_d_vi < f2_d_s):
        f2_ed[j3] = 'Iris-virginica'
        
    elif (f2_d_s < f2_d_vc) and (f2_d_s < f2_d_vi):
        f2_ed[j3] = 'Iris-setosa'
    j3 = j3 + 1


#Accuracy
f2_c = 0
for i in range(0, 45):
    if df_test.Species.values[i] == f2_ed[i]:
        f2_c = f2_c + 1
f2_a = (f2_c / 45) * 100



##Equal Frequency discretization with n = 3 intervals
#Discretizing values
f3_dpl = pd.qcut(pl, q=3, labels=[1,2,3])
f3_dpw = pd.qcut(pw, q=3, labels=[1,2,3])


#Calculating the average
f3_c_vc = 0
f3_c_vi = 0
f3_c_s = 0

f3_spl_vc = 0
f3_spl_vi = 0
f3_spl_s = 0

f3_spw_vc = 0
f3_spw_vi = 0
f3_spw_s = 0

for i in range(0, 105):
    if df.Species.values[i] == 'Iris-versicolor':
        f3_c_vc = f3_c_vc + 1
        f3_spl_vc = f3_spl_vc + f3_dpl.values[i]
        f3_spw_vc = f3_spw_vc + f3_dpl.values[i]
        
    elif df.Species.values[i] == 'Iris-virginica':
        f3_c_vi = f3_c_vi + 1
        f3_spl_vi = f3_spl_vi + f3_dpl.values[i]
        f3_spw_vi = f3_spw_vi + f3_dpl.values[i]
    
    elif df.Species.values[i] == 'Iris-setosa':
        f3_c_s = f3_c_s + 1
        f3_spl_s = f3_spl_s + f3_dpl.values[i]
        f3_spw_s = f3_spw_s + f3_dpl.values[i]
        
f3_apl_vc = f3_spl_vc / f3_c_vc
f3_apl_vi = f3_spl_vi / f3_c_vi
f3_apl_s = f3_spl_s / f3_c_s

f3_apw_vc = f3_spw_vc / f3_c_vc
f3_apw_vi = f3_spw_vi / f3_c_vi
f3_apw_s = f3_spw_s / f3_c_vc


#The Euclidean distance
f3_ed = ['null'] * 45
j4 = 0

for i in range(105, 150):
    #d = ((((x2 - x1 )**2) + ((y2-y1)**2) )**0.5)
    f3_d_vc = ((((f3_apl_vc - f3_dpl.values[i] )**2) + ((f3_apw_vc - f3_dpw.values[i])**2) )**0.5)
    f3_d_vi = ((((f3_apl_vi - f3_dpl.values[i] )**2) + ((f3_apw_vi - f3_dpw.values[i])**2) )**0.5)
    f3_d_s = ((((f3_apl_s - f3_dpl.values[i] )**2) + ((f3_apw_s - f3_dpw.values[i])**2) )**0.5)
    
    if (f3_d_vc < f3_d_vi) and (f3_d_vc < f3_d_s):
        f3_ed[j4] = 'Iris-versicolor'
        
    elif (f3_d_vi < f3_d_vc) and (f3_d_vi < f3_d_s):
        f3_ed[j4] = 'Iris-virginica'
        
    elif (f3_d_s < f3_d_vc) and (f3_d_s < f3_d_vi):
        f3_ed[j4] = 'Iris-setosa'
    j4 = j4 + 1


#Accuracy
f3_c = 0
for i in range(0, 45):
    if df_test.Species.values[i] == f3_ed[i]:
        f3_c = f3_c + 1
f3_a = (f3_c / 45) * 100



##Output of the accuracy percentages
print ('The accuracy percentage when using Equal Interval discretization with n = 2 intervals is: ', i2_a, '%')
print ('The accuracy percentage when using Equal Interval discretization with n = 3 intervals is: ', i3_a, '%')
print ('The accuracy percentage when using Equal Frequency discretization with n = 2 intervals is: ', f2_a, '%')
print ('The accuracy percentage when using Equal Frequency discretization with n = 3 intervals is: ', f3_a, '%')
```
